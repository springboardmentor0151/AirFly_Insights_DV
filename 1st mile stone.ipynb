```python
df = df.drop_duplicates()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python

df['Nationality'].fillna(df['Nationality'].mode()[0], inplace=True)
```

```python
df['Airport Name'].fillna(df['Airport Name'].mode()[0], inplace=True)
```

```python
df['Departure Date'] = pd.to_datetime(df['Departure Date'], format='mixed')
```

```python
df['Month'] = df['Departure Date'].dt.month
```

```python
df['Departure Date'] = pd.to_datetime(df['Departure Date'], format='mixed', errors='coerce')
```

```python
df['Day'] = df['Departure Date'].dt.day
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df['day_of_week'] = df['Departure Date'].dt.dayofweek
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
