```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\prakash\OneDrive\Desktop\Air fly\cleaned_dataset.csv")
```

```python
top_airports = df['Airport Name'].value_counts().head(10)

plt.figure()
top_airports.plot(kind='bar')
plt.title("Top Airports by Number of Flights")
plt.xlabel("Airport Name")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
status_counts = df['Flight Status'].value_counts()

plt.figure()
status_counts.plot(kind='bar')
plt.title("Flight Status Distribution")
plt.xlabel("Flight Status")
plt.ylabel("Number of Flights")
plt.show()
```

```python
month_counts = df['Month'].value_counts().sort_index()

plt.figure()
month_counts.plot(kind='bar')
plt.title("Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

```python
plt.figure()
sns.countplot(x='day_of_week', data=df)
plt.title("Flights by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Flights")
plt.show()
```

```python
plt.figure()
plt.hist(df['Age'], bins=20)
plt.title("Passenger Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Gender', y='Age', data=df)
plt.title("Age Distribution by Gender")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Month'], df['Age'])
plt.title("Passenger Age vs Departure Month")
plt.xlabel("Month")
plt.ylabel("Age")
plt.show()
```

```python
flights_by_month = df.groupby('Month').size()

plt.figure()
flights_by_month.plot(kind='line', marker='o')
plt.title("Number of Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
top_airports = df['Airport Name'].value_counts().head(10)

plt.figure()
top_airports.plot(kind='bar')
plt.title("Top Airports by Number of Flights")
plt.xlabel("Airport Name")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
continent_counts = df['Continents'].value_counts()

plt.figure()
continent_counts.plot(kind='bar')
plt.title("Flights by Continent")
plt.xlabel("Continent")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.show()
```
