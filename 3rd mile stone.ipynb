```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\prakash\OneDrive\Desktop\Air fly\cleaned_dataset.csv")
```

```python
df['Route'] = df['Airport Name'] + " - " + df['Arrival Airport']
```

```python
top_routes = df['Route'].value_counts().head(10)

plt.figure()
top_routes.plot(kind='bar')
plt.title("Top 10 Flight Routes")
plt.xlabel("Route")
plt.ylabel("Number of Flights")
plt.show()
```

```python
pivot_table = df.pivot_table(
    values='Passenger ID',
    index='Airport Name',
    columns='Arrival Airport',
    aggfunc='count'
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, cmap="coolwarm")
plt.title("Flight Routes Heatmap")
plt.xlabel("Arrival Airport")
plt.ylabel("Departure Airport")
plt.show()
```

```python
busiest_airports = df['Airport Name'].value_counts().head(10)

plt.figure()
busiest_airports.plot(kind='bar')
plt.title("Busiest Airports")
plt.xlabel("Airport Name")
plt.ylabel("Number of Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
cancelled = df[df['Flight Status'] == 'Cancelled']

cancel_month = cancelled.groupby('Month').size()

plt.figure()
cancel_month.plot(kind='line', marker='o')
plt.title("Monthly Flight Cancellations")
plt.xlabel("Month")
plt.ylabel("Number of Cancelled Flights")
plt.show()
```

```python
status_counts = df['Flight Status'].value_counts()

plt.figure()
status_counts.plot(kind='bar')
plt.title("Flight Status Distribution")
plt.xlabel("Flight Status")
plt.ylabel("Number of Flights")
plt.show()
```
