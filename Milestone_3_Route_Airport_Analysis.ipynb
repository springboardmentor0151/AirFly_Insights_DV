```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("airline_operations.csv")

df['delay_rate'] = df['arr_del15'] / df['arr_flights']
df['cancel_rate'] = df['arr_cancelled'] / df['arr_flights']
```

Top 10 Busiest Airports

```python
airport_traffic = df.groupby('airport_name')['arr_flights'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

airport_traffic.plot(kind='bar')

plt.title("Top 10 Busiest Airports")
plt.xlabel("Airport")
plt.ylabel("Total Flights")

plt.show()
```

Delay Heatmap by Airline

```python
delay_columns = [
'carrier_delay',
'weather_delay',
'nas_delay',
'security_delay',
'late_aircraft_delay'
]

delay_heatmap = df.groupby('carrier_name')[delay_columns].mean()

plt.figure(figsize=(12,6))

sns.heatmap(delay_heatmap, cmap="YlOrRd")

plt.title("Delay Causes by Airline")

plt.show()
```

Airport Delay Analysis

```python
airport_delay = df.groupby('airport_name')['delay_rate'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

airport_delay.plot(kind='bar', color='orange')

plt.title("Airports with Highest Delay Rate")
plt.xlabel("Airport")
plt.ylabel("Average Delay Rate")

plt.show()
```

Monthly Cancellation Trend

```python
monthly_cancel = df.groupby('month')['cancel_rate'].mean()

plt.figure(figsize=(8,5))

monthly_cancel.plot(marker='o')

plt.title("Monthly Cancellation Trend")
plt.xlabel("Month")
plt.ylabel("Cancellation Rate")

plt.show()
```

Winter Month Analysis

```python
winter_data = df[df['month'].isin([12,1,2])]

winter_cancel = winter_data['cancel_rate'].mean()

print("Average Winter Cancellation Rate:", winter_cancel)
```

```python
winter_months = df[df['month'].isin([12,1,2])]

winter_months.groupby('month')['cancel_rate'].mean().plot(kind='bar')

plt.title("Winter Cancellation Trend")

plt.show()
```
