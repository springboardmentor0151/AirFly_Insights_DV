--CONVERTING DATATYPES 
F_DATE,CATEGORICAL DATA, NUMERICAL_BINARY COLUMN, KEEPING FLOAT FOR DELAY COLS,

```python
df['FL_DATE']=pd.to_datetime(df['FL_DATE'])
df['FL_DATE'].dtype
```

```python
categorical_cols = [
    'OP_CARRIER',
    'ORIGIN',
    'DEST',
    'CANCELLATION_CODE'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
df['CANCELLED'] = df['CANCELLED'].astype('int8')
df['DIVERTED'] = df['DIVERTED'].astype('int8')
```

```python
df['OP_CARRIER_FL_NUM'] = df['OP_CARRIER_FL_NUM'].astype('int32')
```

```python
time_cols = [
    'CRS_DEP_TIME', 'DEP_TIME',
    'WHEELS_OFF', 'WHEELS_ON',
    'CRS_ARR_TIME', 'ARR_TIME'
]

for col in time_cols:
    df[col] = df[col].fillna(0).astype('int32')
```

```python
delay_cols = [
    'DEP_DELAY', 'ARR_DELAY',
    'CARRIER_DELAY', 'WEATHER_DELAY',
    'NAS_DELAY', 'SECURITY_DELAY',
    'LATE_AIRCRAFT_DELAY'
]

for col in delay_cols:
    df[col] = df[col].fillna(0).astype('float32')
```

```python
df['DISTANCE'] = df['DISTANCE'].astype('int32')

elapsed_cols = ['CRS_ELAPSED_TIME', 'ACTUAL_ELAPSED_TIME']
for col in elapsed_cols:
    df[col] = df[col].fillna(0).astype('float32')
```

```python
# CHECKING MEMORY_USAGE
df.info(memory_usage='deep')
```

```python
# EXPLORING SCHEMA TYPES ANS SIZE NULLS
df.isnull().sum()
```

```python
# 1️⃣ Fix TAXI_OUT, TAXI_IN, AIR_TIME (replace null with 0)
time_cols = ['TAXI_OUT', 'TAXI_IN', 'AIR_TIME']
df[time_cols] = df[time_cols].fillna(0)

# 2️⃣ Fix CANCELLATION_CODE properly
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].astype(str)
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].replace('nan', 'Not Cancelled')
df['CANCELLATION_CODE'] = df['CANCELLATION_CODE'].astype('category')
```

FEATURE ENGINEERING AND PREPROCESSING


```python
df['YEAR'] = df['FL_DATE'].dt.year
df['MONTH'] = df['FL_DATE'].dt.month
df['DAY'] = df['FL_DATE'].dt.day
df['DAY_OF_WEEK'] = df['FL_DATE'].dt.dayofweek
```

```python
# CREATED DEPARTURE HOUR 
df['DEP_HOUR'] = df['DEP_TIME'] // 100
```

```python
df['ROUTE'] = df['ORIGIN'].astype(str) + "-" + df['DEST'].astype(str)
```

```python
# season column
def get_season(month):
    if month in [12,1,2]:
        return "Winter"
    elif month in [3,4,5]:
        return "Spring"
    elif month in [6,7,8]:
        return "Summer"
    else:
        return "Fall"

df['SEASON'] = df['MONTH'].apply(get_season)
df['SEASON'] = df['SEASON'].astype('category')
```
