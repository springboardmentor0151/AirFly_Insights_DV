Load Dataset

```python
file_path = "../data/raw/Airline Dataset Updated - v2.csv"
df = pd.read_csv(file_path)

print("Dataset Loaded Successfully ✅")
print("Shape:", df.shape)
```

Handle Missing Values

```python
print("No missing values found ✅")
```

 Convert Departure Date to datetime

```python
print(df['Departure Date'].head())
print(df['Departure Date'].dtype)
```

```python
df['Departure Date'] = pd.to_datetime(
    df['Departure Date'],
    errors='coerce',
    infer_datetime_format=True
)
```

```python
df['Departure Date'].isnull().sum()
```

Feature Engineering

```python
df['Year'] = df['Departure Date'].dt.year
df['Month'] = df['Departure Date'].dt.month
df['Day'] = df['Departure Date'].dt.day
df['DayOfWeek'] = df['Departure Date'].dt.day_name()
```

```python
df.head()
```

Convert Categorical Columns (Memory Optimization)

```python
categorical_cols = [
    'Gender',
    'Nationality',
    'Airport Name',
    'Country Name',
    'Airport Continent',
    'Arrival Airport',
    'Flight Status'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

Save Cleaned Dataset

```python
df.to_csv("../data/processed/cleaned_airline_data.csv", index=False)
print("Milestone 1 Completed ✅")
```
