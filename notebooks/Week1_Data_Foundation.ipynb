```python
import pandas as pd

df = pd.read_csv("../data/flights.csv", nrows=200000)

df.head()
```

```python
df.drop(columns=['index'], inplace=True)
df.head()
```

```python
df.duplicated().sum()
```

```python
df['route'] = df['source_city'] + " → " + df['destination_city']
df.head()
```

```python
categorical_cols = [
    'airline',
    'flight',
    'source_city',
    'departure_time',
    'stops',
    'arrival_time',
    'destination_city',
    'class',
    'route'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')

df.info()
```

```python
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
```
