# Week 1: Data Loading and Preprocessing

## Flight Delay Analysis Project

### Goal

In this notebook, we load the flight dataset and perform the initial data preparation steps. The aim is to understand the structure of the dataset and clean it so it can be used for further analysis.

### What is done in this notebook

* Load the dataset using pandas
* Explore the dataset using `head()`, `info()`, and `describe()`
* Check for missing values
* Convert some columns to categorical to optimize memory
* Create useful new columns like:

  * `date`
  * `month_name`
  * `day_of_week`
  * `hour`
  * `route`
  * `is_delayed`
  * `is_cancelled`
  * `is_weekend`

### Output

After cleaning and preprocessing, the dataset is saved as:

`flights_cleaned_m1.csv`

This cleaned dataset will be used for analysis in the next notebooks.


```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/flights.csv")

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_sample = df.sample(100000, random_state=42)
df_sample.shape
```

```python
categorical_cols = ['carrier', 'origin', 'dest', 'name']

for col in categorical_cols:
    df[col] = df[col].astype('category')

df.info()
```

```python
df[['dep_delay','arr_delay']].isnull().sum()
```

```python
df['is_cancelled'] = df['dep_time'].isnull()
```

```python
df['date'] = pd.to_datetime(df[['year','month','day']])
```

```python
df['month_name'] = df['date'].dt.month_name()
```

```python
df['day_of_week'] = df['date'].dt.day_name()
```

```python
df['hour'] = df['sched_dep_time'] // 100
df.head()
```

```python
df['route'] = df['origin'].astype(str) + "-" + df['dest'].astype(str)
df.head()
```

```python
df[['origin','dest','route']].head()
```

```python
df['is_delayed'] = df['dep_delay'] > 15
```

```python
df['is_weekend'] = df['day_of_week'].isin(['Saturday','Sunday'])
#weekend flag to analyze delay differences between weekdays and weekends.
```

```python
df.duplicated().sum()
```

```python
df.to_csv("../data/flights_cleaned_m1.csv", index=False)
```
