```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")

df = pd.read_csv("../data/flights_cleaned_m1.csv")
df.head()
```

1. Top AIRLINE

```python
top_airlines = df['name'].value_counts().head(10)

plt.figure()
top_airlines.sort_values().plot(kind='barh')
plt.title("Top 10 Airlines by Flight Volume")
plt.xlabel("Number of Flights")
plt.show()
```

2. Top 10 Routes(Bivariate)

```python
top_routes = df['route'].value_counts().head(10)

plt.figure()
top_routes.sort_values().plot(kind='barh')
plt.title("Top 10 Routes by Flight Volume")
plt.xlabel("Number of Flights")
plt.show()
```

3.Busiest Months (Univariate)

```python
monthly_counts = df['month'].value_counts().sort_index()

plt.figure()
monthly_counts.plot(kind='line')
plt.title("Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

4. Flight Distribution by Day of Week

```python
day_counts = df['day_of_week'].value_counts().sort_index()

plt.figure()
day_counts.plot(kind='bar')
plt.title("Flights by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Flights")
plt.show()
```

5. Distribution of Arrival Delay (Histogram)

```python
plt.figure()
df['arr_delay'].hist(bins=50)
plt.title("Distribution of Arrival Delays")
plt.xlabel("Delay (Minutes)")
plt.ylabel("Frequency")
plt.show()
```

6. Boxplot for Delay (Outliers)

```python
plt.figure()
sns.boxplot(x=df['arr_delay'])
plt.title("Boxplot of Arrival Delay")
plt.show()
```

7. Average Delay by Airline (Bivariate)

```python
delay_by_airline = df.groupby('name')['arr_delay'].mean().sort_values()

plt.figure()
delay_by_airline.plot(kind='barh')
plt.title("Average Arrival Delay by Airline")
plt.xlabel("Average Delay (Minutes)")
plt.show()
```

8. Delay by Airport (Origin)

```python
delay_by_airport = df.groupby('origin')['arr_delay'].mean().sort_values()

plt.figure()
delay_by_airport.plot(kind='barh')
plt.title("Average Arrival Delay by Origin Airport")
plt.xlabel("Average Delay (Minutes)")
plt.show()
```

9. Delay by Hour

```python
hour_delay = df.groupby('hour')['arr_delay'].mean()

plt.figure()
hour_delay.plot(kind='line')
plt.title("Average Arrival Delay by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Average Delay (Minutes)")
plt.show()
```

10. Rolling Average

```python
daily_delay = df.groupby('date')['arr_delay'].mean()

rolling_avg = daily_delay.rolling(7).mean()

plt.figure()
daily_delay.plot(label="Daily Avg")
rolling_avg.plot(label="7-Day Rolling Avg")
plt.legend()
plt.title("Daily vs 7-Day Rolling Average Delay")
plt.show()
```

11. Delay Distribution (Histogram)

```python
plt.figure()
df['arr_delay'].hist(bins=50)
plt.title("Distribution of Arrival Delays")
plt.xlabel("Delay (Minutes)")
plt.ylabel("Frequency")
plt.show()
```
