# Week 3: Route, Airport and Seasonal Analysis

### Goal

In this notebook, we go deeper into the dataset to analyze flight delays based on routes, airports, and seasonal patterns.

### What we analyze

* Top routes with the highest number of flights
* Routes with higher delays
* Busiest airports
* Average delay by airport
* Heatmap of delays between airports
* Monthly delay trends
* Weekend vs weekday delays
* Monthly cancellation trends

### Purpose

This analysis helps identify:

* busy flight routes
* airports with higher delays
* seasonal patterns in flight operations


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")

df = pd.read_csv("../data/flights_cleaned_m1.csv")

df.head()
```

1. Top 10 Origin–Destination Routes

```python
top_routes = df['route'].value_counts().head(10)

plt.figure()
top_routes.sort_values().plot(kind='barh')

plt.title("Top 10 Busiest Routes")
plt.xlabel("Number of Flights")
plt.ylabel("Route")

plt.show()
```

2. Delay by Route

```python
route_delay = df.groupby('route')['arr_delay'].mean().sort_values(ascending=False).head(10)

plt.figure()
route_delay.sort_values().plot(kind='barh')

plt.title("Routes with Highest Average Arrival Delay")
plt.xlabel("Average Delay (Minutes)")
plt.ylabel("Route")

plt.show()
```

3. Delay Heatmap by Airport

```python
airport_delay = df.groupby('origin')['arr_delay'].mean().sort_values(ascending=False).head(10)

airport_delay.plot(kind='bar', figsize=(10,6))
plt.title("Airports with Highest Average Delays")
plt.ylabel("Average Delay (minutes)")
plt.xlabel("Airport")
plt.show()
```

4. Busiest Airports

```python
busiest_airports = df['origin'].value_counts().head(10)

plt.figure()

busiest_airports.sort_values().plot(kind='barh')

plt.title("Top 10 Busiest Airports")
plt.xlabel("Number of Flights")
plt.ylabel("Airport")

plt.show()
```

5. Cancellation Trends by Month

```python
cancelled = df[df['is_cancelled'] == True]

monthly_cancel = cancelled.groupby('month').size()

plt.figure()

monthly_cancel.plot(kind='line', marker='o')

plt.title("Monthly Flight Cancellations")
plt.xlabel("Month")
plt.ylabel("Number of Cancelled Flights")

plt.show()
```

6. Cancellation Rate by Airport

```python
cancel_rate_airport = df.groupby('origin')['is_cancelled'].mean().sort_values(ascending=False).head(10)

plt.figure()

cancel_rate_airport.sort_values().plot(kind='barh')

plt.title("Airports with Highest Cancellation Rate")
plt.xlabel("Cancellation Rate")
plt.ylabel("Airport")

plt.show()
```

7. Seasonal Delay Trend

```python
monthly_delay = df.groupby('month')['arr_delay'].mean()

plt.figure()

monthly_delay.plot(kind='line', marker='o')

plt.title("Average Delay by Month")
plt.xlabel("Month")
plt.ylabel("Average Delay (Minutes)")

plt.show()
```

8. Weekend vs Weekday Delays

```python
weekend_delay = df.groupby('is_weekend')['arr_delay'].mean()

plt.figure()

weekend_delay.plot(kind='bar')

plt.title("Weekend vs Weekday Delay")
plt.ylabel("Average Arrival Delay")

plt.xticks([0,1],["Weekday","Weekend"])

plt.show()
```

9. Delay by Route Distance

```python
distance_delay = df.groupby('distance')['arr_delay'].mean()

plt.figure()

plt.scatter(df['distance'], df['arr_delay'], alpha=0.3)

plt.title("Distance vs Arrival Delay")
plt.xlabel("Distance")
plt.ylabel("Arrival Delay")

plt.show()
```
