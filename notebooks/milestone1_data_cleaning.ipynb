```python
df['Departure Date'] = df['Departure Date'].fillna(method='ffill')
df['Flight Status'] = df['Flight Status'].fillna('Unknown')
```

Convert Date Column

```python
df['Departure Date'] = pd.to_datetime(df['Departure Date'], format='mixed', dayfirst=True)
```

```python
# Remove Unnamed columns
df.drop(columns=df.filter(like='Unnamed').columns, inplace=True)

# Handle missing Departure Date
df = df.dropna(subset=['Departure Date'])
df['Departure Date'] = pd.to_datetime(df['Departure Date'], errors='coerce')


# Handle missing Flight Status
df['Flight Status'] = df['Flight Status'].fillna('Unknown')

# Remove duplicates
df.drop_duplicates(inplace=True)

# Final check
df.info()
df.isnull().sum()
```

```python
df['Airport Name'] = df['Airport Name'].astype('category')
df['Arrival Airport'] = df['Arrival Airport'].astype('category')
```

5. Feature Engineering

5.1 Date Features

Visualizations a pie chart for flight status and a bar plot for gender distribution.

Create Useful Features

 Date-Based Features

From Departure Date:

Year

Month

These will help in identifying seasonal or monthly patterns

```python
df['Year'] = df['Departure Date'].dt.year
df['Month'] = df['Departure Date'].dt.month
```

Time-Based Feature

```python
df['Time of Departure'] = pd.to_datetime(df['Time of Departure'], errors='coerce')
df['DepartureHour'] = df['Time of Departure'].dt.hour
```

Route Creation

```python
df['Route'] = df['Airport Name'].astype(str) + " - " + df['Arrival Airport'].astype(str)
```

```python
df['Route'] = df['Route'].astype('category')
```

```python
df['Flight Status'].unique()
```

Delay and Cancellation Indicators

```python
# Clean status column
df['Flight Status'] = df['Flight Status'].astype(str).str.strip()

# Create delay flag
df['IsDelayed'] = (df['Flight Status'].str.lower() == 'delayed').astype(int)

# Create cancellation flag
df['IsCancelled'] = (df['Flight Status'].str.lower() == 'cancelled').astype(int)
```

Memory Optimization (Important for Large Dataset)

Reduce Memory Usage

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype('category')
```
